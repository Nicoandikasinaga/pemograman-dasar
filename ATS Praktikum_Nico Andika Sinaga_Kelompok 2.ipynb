```python
# PROGRAM MONITORING PEMBANGKIT LISTRIK
# Mengimplementasikan percabangan, perulangan, dan fungsi

# ===== FUNGSI UTAMA =====
def hitung_efisiensi(beban_terpakai, kapasitas_maksimum):
    """Fungsi untuk menghitung efisiensi pembangkit"""
    if kapasitas_maksimum <= 0:
        return 0
    return (beban_terpakai / kapasitas_maksimum) * 100

def klasifikasikan_beban(persentase):
    """Fungsi untuk mengklasifikasikan beban berdasarkan persentase"""
    if persentase <= 60:
        return "Beban Ringan"
    elif persentase <= 85:
        return "Beban Sedang"
    else:
        return "Beban Berat"

def cek_status_suhu(suhu):
    """Fungsi untuk mengecek status suhu"""
    if suhu <= 90:
        return "Normal"
    elif suhu < 100:
        return "Warning"
    else:
        return "SHUTDOWN"

# ===== DATA SENSOR =====
data_sensor = {
    'suhu_boiler': [92.5, 88.7, 95.1, 102.3, 97.8],
    'vibrasi_mesin': [1.2, 0.9, 2.5, 3.1, 1.8],
    'beban_unit': [45, 75, 90, 60, 85]  # dalam MW
}

kapasitas_maksimum = 100  # MW

# ===== PROGRAM UTAMA =====
def main():
    print("=" * 50)
    print("   SISTEM MONITORING PEMBANGKIT LISTRIK")
    print("=" * 50)

    # === PERCABANGAN - Input suhu manual ===
    print("\n1. MONITORING SUHU MANUAL")
    try:
        suhu_input = float(input("Masukkan suhu air pendingin (¬∞C): "))
        status_suhu = cek_status_suhu(suhu_input)
        print(f"STATUS: {status_suhu}")
    except ValueError:
        print("Error: Input harus berupa angka!")

    # === PERULANGAN FOR - Monitoring semua sensor ===
    print("\n2. MONITORING SEMUA SENSOR")
    print("-" * 40)

    # Monitoring suhu boiler
    print("\nüîç MONITORING SUHU BOILER:")
    for i, suhu in enumerate(data_sensor['suhu_boiler'], 1):
        status = cek_status_suhu(suhu)
        print(f"   Titik {i}: {suhu}¬∞C - {status}")
        if status == "SHUTDOWN":
            print("      ‚ö†Ô∏è  ALERT: Unit perlu dimatikan!")

    # Monitoring vibrasi mesin
    print("\nüîç MONITORING VIBRASI MESIN:")
    for i, vibrasi in enumerate(data_sensor['vibrasi_mesin'], 1):
        print(f"   Unit {i}: {vibrasi} mm/s", end="")
        if vibrasi > 2.0:
            print(" ‚ö†Ô∏è  Warning: Vibrasi tinggi!")
        else:
            print()

    # Monitoring beban unit
    print("\nüîç MONITORING BEBAN UNIT:")
    for i, beban in enumerate(data_sensor['beban_unit'], 1):
        persentase = hitung_efisiensi(beban, kapasitas_maksimum)
        status_beban = klasifikasikan_beban(persentase)
        print(f"   Unit {i}: {beban} MW ({persentase:.1f}%) - {status_beban}")

    # === PERULANGAN WHILE - Simulasi pemanasan ===
    print("\n3. SIMULASI PROSES PEMANASAN BOILER")
    print("-" * 40)

    suhu_saat_ini = 60
    target_suhu = 85
    iterasi = 1

    while suhu_saat_ini < target_suhu:
        print(f"   Iterasi {iterasi}: Suhu = {suhu_saat_ini}¬∞C")
        suhu_saat_ini += 5
        iterasi += 1

        # Safety check - jika suhu terlalu tinggi
        if suhu_saat_ini >= 100:
            print("   ‚ö†Ô∏è  ALERT: Suhu kritis! Menghentikan pemanasan.")
            break

    print(f"   ‚úÖ Proses pemanasan selesai. Suhu akhir: {suhu_saat_ini}¬∞C")

    # === ANALISIS KINERJA KESELURUHAN ===
    print("\n4. ANALISIS KINERJA KESELURUHAN")
    print("-" * 40)

    # Hitung statistik
    total_beban = sum(data_sensor['beban_unit'])
    efisiensi_rata_rata = hitung_efisiensi(total_beban, kapasitas_maksimum * len(data_sensor['beban_unit']))

    # Count alerts
    suhu_kritis = sum(1 for s in data_sensor['suhu_boiler'] if s >= 100)
    vibrasi_tinggi = sum(1 for v in data_sensor['vibrasi_mesin'] if v > 2.0)
    beban_berat = sum(1 for b in data_sensor['beban_unit'] if hitung_efisiensi(b, kapasitas_maksimum) > 85)

    print(f"   Efisiensi rata-rata: {efisiensi_rata_rata:.1f}%")
    print(f"   Unit dengan suhu kritis: {suhu_kritis}")
    print(f"   Unit dengan vibrasi tinggi: {vibrasi_tinggi}")
    print(f"   Unit dengan beban berat: {beban_berat}")

    # Analisis
    print("\n5. Analisis:")
    print("-" * 40)
    if suhu_kritis > 0 or vibrasi_tinggi > 2:
        print("   ‚ùå Analisis: Lakukan maintenance segera!")
    elif efisiensi_rata_rata > 80:
        print("   ‚úÖ KINERJA: Optimal - Pertahankan operasi")
    else:
        print("   ‚ÑπÔ∏è  KINERJA: Normal - Lanjutkan monitoring")

# ===== MENJALANKAN PROGRAM =====
if __name__ == "__main__":
    main()
```
