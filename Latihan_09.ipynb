```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Nico_andika_sinaga = pd.read_csv('/content/products.csv')
#import file csv
```

```python
# Ringkasan statistik
Nico_andika_sinaga.describe()
```

```python
# Ringkasan statistik
Nico_andika_sinaga.describe()

# Rata-rata harga
rata_rata_customer_id = Nico_andika_sinaga['price'].mean()
print("Rata-rata harga:", rata_rata_customer_id)

# Modus jumlah order_id
modus_order_id = Nico_andika_sinaga['quantity'].mode()
print("Modus size:", modus_order_id)
```

```python
Nico_andika_sinaga['quantity'].hist()
plt.title("Frekuensi jumlah quantity")
plt.xlabel("quantity")
plt.ylabel("frekuensi")
plt.show()
```

```python
# Korelasi
corr_matrix = Nico_andika_sinaga[['quantity', 'price']].corr()
print("Korelasi:", corr_matrix)

# Kovarian
cov_matrix = Nico_andika_sinaga[['quantity', 'price']].cov()
print("Kovarian:", cov_matrix)
```

```python
KOLOM_GROUPING = 'product_type'

group_mean_data = Nico_andika_sinaga.groupby(KOLOM_GROUPING).mean(numeric_only=True)
print(group_mean_data)

group_agg_data = Nico_andika_sinaga.groupby(KOLOM_GROUPING).agg({
    'quantity': ['mean', 'min', 'max'],
    'price': ['mean', 'min', 'max']
})
print(group_agg_data)
```
