```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ringkasan statistik
Nico_andika_sinaga.describe()

# Rata-rata customer_id
rata_rata_customer_id = Nico_andika_sinaga['customer_id'].mean()
print("Rata-rata customer_id:", rata_rata_customer_id)

# Modus jumlah order_id
modus_order_id = Nico_andika_sinaga['order_id'].mode()
print("Modus Order_id:", modus_order_id)
```

```python
import matplotlib.pyplot as plt

Nico_andika_sinaga['order_date'].hist()
plt.title("order_id")
plt.xlabel("customer_id")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# Korelasi
corr_matrix = Nico_andika_sinaga[['customer_id', 'order_id']].corr()
print("Korelasi:n", corr_matrix)

# Kovarian
cov_matrix = Nico_andika_sinaga[['customer_id', 'order_id']].cov()
print("Kovarian:n", cov_matrix)
```

```python
# Menambah kolom kategori
Nico_andika_sinaga['payment'] = Nico_andika_sinaga['order_id'].apply(lambda x: 'Banyak' if x >= 3 else 'Sedikit')
Nico_andika_sinaga

# Rata-rata customer_id per payment
group_mean = Nico_andika_sinaga.groupby('payment')['customer_id'].mean()
print(group_mean)

# Agregasi lengkap
group_agg = Nico_andika_sinaga.groupby('payment').agg({
    'customer_id': ['mean', 'max'],
    'order_id': ['mean', 'count']
})
group_agg
```
