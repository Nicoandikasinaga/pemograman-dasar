```python
# Program menampilkan nama sebanyak 200 kali

# Nama yang akan ditampilkan
nama = "Nico Andika Sinaga"

# Menggunakan perulangan for
print("Menggunakan perulangan for:")
for i in range(200):
    print(f"{i+1}. {nama}")

print("\n" + "="*50 + "\n")

# Menggunakan perulangan while
print("Menggunakan perulangan while:")
counter = 1
while counter <= 200:
    print(f"{counter}. {nama}")
    counter += 1
```

```python
# Program menampilkan angka 1-500 menggunakan while

print("Angka 1 sampai 500 menggunakan while:")
angka = 1
while angka <= 500:
    print(f"{angka:3d}", end=" ")
    if angka % 15 == 0:  # Ganti baris setiap 15 angka
        print()
    angka += 1
```

```python
# Menampilkan deret bilangan ganjil dari 1 sampai 200 menggunakan while

# Inisialisasi variabel
angka = 1

# Perulangan while
while angka <= 200:
    # Cek apakah angka ganjil
    if angka % 2 != 0:
        print(angka, end=" ")
    angka += 1
```

```python
print("\nSOAL : BILANGAN PRIMA")
print("=" * 40)

batas = int(input("Masukkan batas atas untuk mencari bilangan prima: "))

print(f"\nBilangan prima dari 1 sampai {batas}:")

# Menggunakan for loop
print("Menggunakan for loop:")
for num in range(2, batas + 1):
    is_prima = True
    for i in range(2, int(num ** 0.5) + 1):
        if num % i == 0:
            is_prima = False
            break
    if is_prima:
        print(num, end=" ")

print("\n\nMenggunakan while loop:")
# Menggunakan while loop
num = 2
while num <= batas:
    is_prima = True
    i = 2
    while i <= int(num ** 0.5):
        if num % i == 0:
            is_prima = False
            break
        i += 1
    if is_prima:
        print(num, end=" ")
    num += 1
```
